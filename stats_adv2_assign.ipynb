1. What is hypothesis testing in statistics

Hypothesis testing in statistics is a method used to make decisions or draw conclusions about a population based on sample data.
* it is likely to make assumptions about population parameter to draw conclusions either it is true or not.

2. What is the null hypothesis, and how does it differ from the alternative hypothesis

The null hypothesis (H₀) assumes no effect, no difference, or no relationship between variables . The alternative hypothesis (H₁) suggests there is an effect, difference, or relationship.

The key difference is that H₀ is the default assumption, while H₁ challenges it. Hypothesis testing determines whether to reject H₀ in favor of H₁.

1. example : About the new drug in the market 
        H0 : "the new drug has no effect"
        H1 : "the new drug improves recovery"


3. What is the significance level in hypothesis testing, and why is it important

The significance level (α) in hypothesis testing is the threshold for deciding whether to reject the null hypothesis. 
It represents the maximum probability of making a Type I error (rejecting H₀ when it’s true). Common values are 0.05 (5%) or 0.01 (1%).

It’s important because it controls how strict the test is, balancing the risk of false conclusions and ensuring reliable results.

4. What does a P-value represent in hypothesis testing

A p-value in hypothesis testing represents the probability of obtaining results as extreme as, or more extreme than, the observed data, assuming the null hypothesis (H₀) is true.

In simple terms:

A small p-value (e.g., < 0.05) suggests the observed result is unlikely under H₀, so you may reject H₀.
A large p-value (e.g., > 0.05) suggests the observed result is consistent with H₀, so you fail to reject H₀.

5. How do you interpret the P-value in hypothesis testing

To interpret the p-value in hypothesis testing:

Small p-value:(0.05)

Reject the null hypothesis (H₀).
This means the observed data is unlikely to happen if H₀ is true, so there's evidence for the alternative hypothesis (H₁).
Example: p-value = 0.03 → Since 0.03 < 0.05, you reject H₀.
Large p-value:

Fail to reject the null hypothesis (H₀).
This means the observed data is not very unusual, and there's not enough evidence to support H₁.
Example: p-value = 0.08 → Since 0.08 > 0.05, you fail to reject H₀.

6. What are Type 1 and Type 2 errors in hypothesis testing

* Type 1 Error = False Positive: You wrongly reject H₀ (saying there's an effect when there isn't).
Ex : Accepting that the person has the disease when actually not

* Type 2 Error = False Negative: You wrongly fail to reject H₀ (missing an effect when there is one).
Ex : A test wrongly tells you that the person does not have the disease when they actually do. This happens when the test misses a real effect.

1. Write a Python program to perform a Z-test for comparing a sample mean to a known population mean and
interpret the results

```python
def z_test(sample_data,pop_mean,pop_std,los):
    ''' 
    To Perform the Z_test we need to have sample data, 
    find the mean of the data
    find z_statistics
    parameter:
    Takes 4 parameter
    
    Returns None: prints the results and interpretation
    '''
    # calculate sample mean
    samp_mean = np.mean(sample_data)
    sample_size = len(sample_data)
    
    # calculate standard error
    std_error = pop_std / np.sqrt(sample_size)
    
    # calculate z_statistics
    z_stats = (samp_mean - pop_mean) / std_error
    
    # calculate p-value
    p_value = 2 * (1 - stats.norm.cdf(z_stats))
    
    if p_value < los:
        print("Reject H0")
        print("There is a significant difference between sample mean and population mean")
    else:
        print("Fail to reject H0")
        print("There is no significant difference between sample mean and population mean")

    

# Sample data
sample_data = [15, 16, 14, 15, 14, 17, 18, 15, 14, 16]
pop_mean = 15.5
pop_std = 2.0
los = 0.05

# Perform Z-test
z_test(sample_data,pop_mean,pop_std,los)

```

4. Perform a two-tailed Z-test using Python and visualize the decision region on a plot

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats

# Step 1: Simulate random data (e.g., a sample from a population)
np.random.seed(42)  # For reproducibility
sample_data = np.random.normal(loc=50, scale=10, size=100)  # Mean=50, Std=10, n=100

# Step 2: Set the hypothesized population mean and population standard deviation
population_mean = 50
population_std = 10  # Known population standard deviation
sample_size = len(sample_data)

# Step 3: Calculate the sample mean
sample_mean = np.mean(sample_data)

# Step 4: Compute the Z-statistic
z_statistic = (sample_mean - population_mean) / (population_std / np.sqrt(sample_size))

# Step 5: Calculate the p-value (for a two-tailed test)
p_value = 2 * (1 - stats.norm.cdf(abs(z_statistic)))

# Step 6: Output the results
print(f"Z-statistic: {z_statistic}")
print(f"P-value: {p_value}")

# Step 7: Visualization of the decision region
alpha = 0.05  # Significance level
critical_value = stats.norm.ppf(1 - alpha / 2)  # For two-tailed test, critical value at 0.025 and 0.975 quantiles

# Create the Z-distribution
x = np.linspace(-5, 5, 1000)
y = stats.norm.pdf(x)

# Plot the Z-distribution curve
plt.plot(x, y, label='Standard Normal Distribution')

# Highlight the rejection regions (critical regions)
plt.fill_between(x, y, where=(x <= -critical_value) | (x >= critical_value), color='red', alpha=0.5, label='Rejection Region')

# Highlight the acceptance region
plt.fill_between(x, y, where=(x > -critical_value) & (x < critical_value), color='green', alpha=0.5, label='Acceptance Region')

# Mark the Z-statistic on the plot
plt.axvline(z_statistic, color='blue', linestyle='--', label=f'Z-statistic: {z_statistic:.2f}')
plt.axvline(-z_statistic, color='blue', linestyle='--')

# Labels and Title
plt.title("Two-Tailed Z-Test: Decision Region")
plt.xlabel("Z-Statistic")
plt.ylabel("Density")
plt.legend()

# Show the plot
plt.show()

# Interpretation
if p_value < alpha:
    print("Reject the null hypothesis (sample mean is significantly different from the population mean).")
else:
    print("Fail to reject the null hypothesis (sample mean is not significantly different from the population mean).")
```

5. Create a Python function that calculates and visualizes Type 1 and Type 2 errors during hypothesis testing

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats

def visualize_errors(population_mean, population_std, sample_size, alpha=0.05, true_effect=None):
    # Step 1: Simulate data under the null hypothesis
    np.random.seed(42)
    null_data = np.random.normal(loc=population_mean, scale=population_std, size=sample_size)
    
    # Step 2: Simulate data under the alternative hypothesis (if there is an effect)
    if true_effect is not None:
        alternative_data = np.random.normal(loc=population_mean + true_effect, scale=population_std, size=sample_size)
    else:
        alternative_data = null_data  # No effect if true_effect is None
    
    # Step 3: Perform Z-test under both null and alternative hypothesis
    # Calculate Z-statistic for null hypothesis
    sample_mean_null = np.mean(null_data)
    z_statistic_null = (sample_mean_null - population_mean) / (population_std / np.sqrt(sample_size))
    
    # Calculate Z-statistic for alternative hypothesis (if true effect is present)
    sample_mean_alt = np.mean(alternative_data)
    z_statistic_alt = (sample_mean_alt - population_mean) / (population_std / np.sqrt(sample_size))
    
    # Step 4: Calculate critical Z-value for two-tailed test (α=0.05)
    critical_value = stats.norm.ppf(1 - alpha / 2)
    print(z_statistic_null)
    print(z_statistic_alt)
    print(critical_value)
    
    # Step 5: Calculate Type I and Type II errors
    # Type I error: Rejecting null hypothesis when it's true
    type_1_error = stats.norm.cdf(-critical_value) + (1 - stats.norm.cdf(critical_value))
    
    # Type II error: Failing to reject null hypothesis when it's false
    type_2_error = stats.norm.cdf(critical_value - z_statistic_alt) - stats.norm.cdf(-critical_value - z_statistic_alt)
    
    # Step 6: Visualization of Type I and Type II errors
    x = np.linspace(-5, 5, 1000)
    y = stats.norm.pdf(x)
    
    # Plot the normal distribution
    plt.plot(x, y, label="Standard Normal Distribution")
    
    # Mark the rejection regions (Type I Error)
    plt.fill_between(x, y, where=(x <= -critical_value) | (x >= critical_value), color='red', alpha=0.5, label="Rejection Region (Type I Error)")
    
    # Mark the acceptance region
    plt.fill_between(x, y, where=(-critical_value < x) & (x < critical_value), color='green', alpha=0.5, label="Acceptance Region")
    
    # Mark the Z-statistics
    plt.axvline(z_statistic_null, color='blue', linestyle='--', label=f'Z-statistic (Null): {z_statistic_null:.2f}')
    plt.axvline(z_statistic_alt, color='orange', linestyle='--', label=f'Z-statistic (Alternative): {z_statistic_alt:.2f}')
    
    # Add labels, legend, and title
    plt.title("Type I and Type II Errors in Hypothesis Testing")
    plt.xlabel("Z-Statistic")
    plt.ylabel("Density")
    plt.legend()
    
    # Show the plot
    plt.show()
    
    # Print the calculated errors
    print(f"Type I Error (α): {type_1_error:.4f}")
    print(f"Type II Error (β): {type_2_error:.4f}")

# Example usage
population_mean = 50
population_std = 10
sample_size = 100
alpha = 0.05
true_effect = 2  # Simulating a scenario where the true mean is 2 units higher than the population mean

visualize_errors(population_mean, population_std, sample_size, alpha, true_effect)
```

9. Write a Python function to calculate the confidence interval for a sample mean and explain its significance

```python
def cal_conf_intr(sample,confidence_level=0.95):
    """
    Calculate confidence unterval
    parameters:
    sample
    confidence level
    
    output: Confidence interval values
    """
    n = len(sample)
    sample_std = np.std(sample, ddof=1)
    mean = np.mean(sample)
    
    standard_error = sample_std/np.sqrt(n)
    # to find critical value
    critical_value = stats.norm.ppf((1+confidence_level)/2)
    
    # to find the margin error
    margin_error = critical_value * standard_error
    
    # the lower and upper intervals are
    lower_interval = mean - margin_error
    upper_interval = mean + margin_error
    return lower_interval,upper_interval
data = [23, 27, 25, 30, 24, 26, 28, 22, 29, 31]
lower,upper = cal_conf_intr(data,confidence_level=0.95)
print(f" The Confidence interval is ({lower.round(2)},{upper.round(2)})")
```

11. Implement a Bayesian inference method using Bayes' Theorem in Python and explain the process

```python
def posterior_probability(prior_prob,likelihood_B_given_A,likelihood_B_given_not_A):
    prior_not_true = 1-prior_prob
    p_evidence = (prior_prob * likelihood_B_given_A) + (prior_not_true * likelihood_B_given_not_A)
    # posteriro probability
    p_h_given_e = (likelihood_B_given_A * prior_prob) / p_evidence
    return p_h_given_e
# probability of A has disease before test
prior_prob = 0.01

# probab of test is positive given the person has the disease
likelihood_B_given_A = 0.95

# probab if test is positive but the person does not has disease
likelihood_B_given_not_A = 0.05

result = posterior_probability(prior_prob,likelihood_B_given_A,likelihood_B_given_not_A)
print(f"The probability of having the disease is {result}")
```

24. Perform a hypothesis test for population variance using a Chi-square distribution and interpret the results

```python
import numpy as np
from scipy import stats

# Step 1: Simulate some data
data = np.random.normal(loc=50, scale=10, size=30)  # Simulate 30 data points

# Step 2: Set the hypothesized population variance
population_variance = 100  # Hypothesized population variance (sigma_0^2)

# Step 3: Calculate the sample variance
sample_variance = np.var(data, ddof=1)  # Sample variance (using ddof=1 for sample variance)

# Step 4: Calculate the Chi-Square statistic
n = len(data)  # Sample size
chi_square_stat = (n - 1) * sample_variance / population_variance  # Test statistic

# Step 5: Degrees of freedom (n - 1)
df = n - 1  # Degrees of freedom

# Step 6: Find the critical values using Chi-Square distribution
alpha = 0.05  # Significance level (5%)
critical_value_lower = stats.chi2.ppf(alpha / 2, df)  # Lower critical value
critical_value_upper = stats.chi2.ppf(1 - alpha / 2, df)  # Upper critical value

# Step 7: Compare the test statistic with the critical values
print(f"Chi-square statistic: {chi_square_stat}")
print(f"Critical value lower: {critical_value_lower}")
print(f"Critical value upper: {critical_value_upper}")

# Step 8: Make decision based on comparison
if chi_square_stat < critical_value_lower or chi_square_stat > critical_value_upper:
    print("Reject the null hypothesis: The population variance is significantly different from the hypothesized value.")
else:
    print("Fail to reject the null hypothesis: The population variance is not significantly different from the hypothesized value.")
```

26. Implement an F-test for comparing the variances of two datasets, then interpret and visualize the results

```python
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt

# Simulated data: Two sample datasets
data1 = np.random.normal(50, 10, 100)  # Sample 1: mean = 50, std = 10
data2 = np.random.normal(60, 15, 100)  # Sample 2: mean = 60, std = 15

# Calculate the variances of the two datasets
var1 = np.var(data1, ddof=1)
var2 = np.var(data2, ddof=1)

# Perform F-test: H0: var1 = var2, H1: var1 != var2
f_stat = var1 / var2 if var1 > var2 else var2 / var1  # F-statistic (larger variance over smaller variance)
dfn = len(data1) - 1  # Degrees of freedom for the numerator
dfd = len(data2) - 1  # Degrees of freedom for the denominator

# Calculate the p-value for the F-statistic
p_value = 2 * min(stats.f.cdf(f_stat, dfn, dfd), 1 - stats.f.cdf(f_stat, dfn, dfd))

# Output the results
print(f"F-statistic: {f_stat}")
print(f"P-value: {p_value}")

# Interpretation
alpha = 0.05
if p_value < alpha:
    print("Reject the null hypothesis: The variances are significantly different.")
else:
    print("Fail to reject the null hypothesis: The variances are not significantly different.")

# Visualize the data distributions with boxplots
plt.figure(figsize=(8, 6))
plt.boxplot([data1, data2], labels=['Data1', 'Data2'])
plt.title('Boxplot Comparison of Variances')
plt.show()
```
