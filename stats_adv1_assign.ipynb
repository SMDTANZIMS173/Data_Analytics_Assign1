13. What is the Central Limit Theorem (CLT), and why is it critical in statistics

The Central Limit Theorem (CLT) states that when you take multiple random samples from any population, no matter the shape of the original population's distribution, the distribution of the sample means will be approximately normal (bell-shaped), as long as the sample size is large enough (typically n≥30).

* It helps transform complex real-world data into a form that is easier to analyze and interpret, especially when working with averages and making predictions based on sample data.


14. How does the Central Limit Theorem relate to the normal distribution

The Central Limit Theorem explains that the distribution of sample means will always approximate a normal distribution (bell curve), regardless of the shape of the original data, as long as the sample size is sufficiently large. 
* This makes the normal distribution an important tool for statistical analysis and decision-making in a wide range of fields.

16. How do you calculate a Z-score, and what does it represent


The Z-score is a standardized score that represents how many standard deviations a data point (or sample statistic) is away from the mean of a distribution. It allows us to compare data points from different distributions on the same scale, making it easier to identify outliers or determine the relative position of data points.

Formula for Z-score:
Z= (X−μ)/σ
​
Where:

X = the individual data point or sample mean.
μ (mu) = the population mean.
σ (sigma) = the population standard deviation.
It is used in standard normal distribution

* A positive Z-score means the data point is above the mean, and a negative Z-score means it is below the mean.

17. What are point estimates and interval estimates in statistics

1. Point Estimates:
A point estimate is a single value that is used to estimate an unknown population parameter. It is the best guess based on the available data, but it does not provide any information about how precise that estimate is.

2. An interval estimate provides a range of values (interval) within which the population parameter is likely to lie, along with an associated level of confidence. This is a more informative approach because it accounts for variability and uncertainty in the estimate.
* Interval estimates are generally preferred because they provide more information about the accuracy of the estimate.

18. What is the significance of confidence intervals in statistical analysis


* Confidence intervals (CIs) play a crucial role in statistical analysis by providing a range of values that is likely to contain the true population parameter, along with a specified level of confidence. They give more information than just a point estimate, which is just a single value.

1. Quantify Uncertainty: Confidence intervals provide a range of values and reflect the uncertainty of an estimate.
2. Reliability: The confidence level tells you how likely the interval is to contain the true population parameter.
3. Informed Decision-Making: They guide decision-making in areas like business, medicine, and research.
4. Statistical Significance: Confidence intervals help in determining whether observed results are statistically significant.
5. Avoid Over-Reliance on Point Estimates: They offer a more comprehensive view by giving a range instead of a single point estimate.


19. What is the relationship between a Z-score and a confidence interval

1. The Z-score corresponds to the critical value used to calculate the confidence interval.
2. It defines how many standard deviations away the true population parameter is likely to be from the sample mean, given a specified confidence level.
3. A larger Z-score (higher confidence level) results in a wider confidence interval, indicating more uncertainty.
4. The relationship between Z-scores and confidence intervals is key in understanding how we estimate population parameters with a specified degree of certainty.

21. What are the assumptions for applying the Central Limit Theorem

* The Central Limit Theorem (CLT) assumes that the data is randomly sampled, and the sample size is large enough (typically 30 or more). 
* It also assumes that the samples are independent and that the population has a finite mean and variance. 
* These conditions help the sampling distribution of the sample mean approach a normal distribution, even if the original population is not normal.

22. What is the concept of expected value in a probability distribution

The **expected value** in a probability distribution is the average or mean value you would expect to get if you repeated an experiment infinitely many times. It is calculated by multiplying each possible outcome by its probability and summing up the results. 

**Formula:**  
E(X) = sum (xi.P(xi))

Where (xi) is each outcome and (P(xi)) is its probability.

23. How does a probability distribution relate to the expected outcome of a random variable

A probability distribution shows the likelihood of all possible outcomes of a random variable. The expected outcome (or expected value) is the weighted average of these outcomes, where each outcome is weighted by its probability. It represents the long-term average result if the experiment is repeated many times.

6. Write a Python program to calculate and plot the cumulative distribution function (CDF) of a discrete
uniform distribution

```python
import numpy as np
import matplotlib.pyplot as plt

# Step 1: Define parameters for the discrete uniform distribution
low = 1  # The lowest value of the distribution
high = 6  # The highest value of the distribution
n = 1000  # Number of random samples to generate

# Step 2: Generate the samples from the discrete uniform distribution
samples = np.random.randint(low, high + 1, n)

# Step 3: Calculate the CDF
# Get unique values and their counts in the samples
unique_values, counts = np.unique(samples, return_counts=True)

# Calculate the cumulative sum of the counts
cumulative_counts = np.cumsum(counts)

# Normalize to get the CDF (divide by the total number of samples)
cdf = cumulative_counts / n

# Step 4: Plot the CDF
plt.step(unique_values, cdf,  where='post',label="CDF", color='blue', linewidth=2)

# Adding labels and title
plt.title('Cumulative Distribution Function (CDF) of Discrete Uniform Distribution')
plt.xlabel('Value')
plt.ylabel('CDF')
plt.grid(True)
plt.xticks(unique_values)
plt.yticks(np.arange(0, 1.1, 0.1))
plt.legend()
# Display the plot
plt.show()
```

15. Implement hypothesis testing using Z-statistics for a sample dataset

```python
from scipy.stats import norm
import numpy as np

# Sample data
sample_data = [85, 90, 88, 92, 87, 93, 86, 91, 89, 94]
population_mean = 90
pop_std = 5
sample_size = len(sample_data)

sample_mean = np.mean(sample_data)

#calculate z-stat
z_stat = (sample_mean - population_mean) / (pop_std/np.sqrt(sample_size))

# calculate p-value
p_value = 2 * (1-norm.cdf(abs(z_stat)))

#LOS
alpha =0.05

if p_value < alpha:
    print("Reject H0")
else:
    print("Failed to reject H0")
```

16. Create a confidence interval for a dataset using Python and interpret the result

```python
from scipy.stats import norm
import numpy as np

# Sample data
sample_data = [85, 90, 88, 92, 87, 93, 86, 91, 89, 94]
population_mean = 90
pop_std = 5
sample_size = len(sample_data)

sample_mean = np.mean(sample_data)

#calculate z-stat
z_stat = (sample_mean - population_mean) / (pop_std/np.sqrt(sample_size))
print(z_stat)

# 95% CI
z_critical = norm.ppf(0.975)
# print(z_critical)
margin_error = z_critical * (pop_std/np.sqrt(sample_size))
ci_lower = sample_mean - margin_error
ci_higher = sample_mean + margin_error
print(f"Confidence interval is {ci_lower},{ci_higher}")
```
